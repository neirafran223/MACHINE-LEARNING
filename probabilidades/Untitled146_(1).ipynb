# Distribucion binomial

```python
n, p = 120, 0.45
x = np.arange(ss.binom.ppf(0.01, n, p),ss.binom.ppf(0.99, n, p))
# Se definen la funci´on de masa de probabilidad y acumulada
fmp = ss.binom.pmf(x, n, p) # Funcion de masa de probabilidad
fda = ss.binom.cdf(x, n, p) # Funcion de masa de probabilidad acumulada

plt.subplots(1)
plt.vlines(x, 0, fmp, colors='b', lw=8, alpha=.6) # gr´afico de barras.
plt.xlim(40,68)
plt.ylim(0,0.14)

plt.subplots(1)
plt.vlines(x, 0, fda, colors='b', lw=8, alpha=.6) # gr´afico de barras.
plt.xlim(40,68)
plt.ylim(0,1)
```

En un estudio piloto se encontro que a 80 % de personas le agradó el sabor de una sopa instantanea que se va a lanzar en el mercado ¿Cual es la probabilidad de que, en una muestra de 15 personas le agrade el sabor de la nueva sopa?

1) 12 personas

2) 5 personas

3) en una muestra de 50 personas le desagrade el sabor de la nueva sopa a 40 de ellas

```python
# X: cantidad de personas que le gusta la sopa
# n = 15, p = 0.8

p5 = ss.binom(n = 15, p = 0.8).pmf(12)# funcion de cuantia
print(f"La probabilidad de que le guste a 12 personas es de: {p5}")
print()
p6 = ss.binom(15, 0.8).pmf(5)
print(f"La probabilidad de que le guste a 5 personas es de: {p6}")
print()
lista = [k for k in range(1,16)]
probabilidades = []
for k in lista:
    print(f"la probabilidad de que le guste a {k} es de {ss.binom(15,0.8).pmf(k)}")
    probabilidades.append(ss.binom(15,0.8).pmf(k))

print()
p7 = ss.binom(40, 0.2).pmf(40)
print(f"la probabilidad de que no le guste a 40 personas es de: {p7}")
```

# Distribucion de poisson

```python
# Graficando
mu = 50 # parametro de forma
poisson = ss.poisson(mu) # Distribuci´on
x = np.arange(poisson.ppf(0.001),poisson.ppf(0.999),1)
fmp = poisson.pmf(x) # Funci´on de Masa de Probabilidad
plt.vlines(x, 0, fmp, colors='b', lw=8, alpha=.6)
# Arreglos visuales
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.xlim(25,65)
plt.ylim(0,0.08)
plt.title('Funcion de Distribucion de Poisson')
plt.ylabel('Probailidad')
plt.xlabel('Valores')
plt.show()
```

```python
fdc = ss.poisson(mu).cdf(x)
plt.vlines(x, 0, fdc, colors='b', lw=8, alpha=.6)
# Arreglos visuales
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.xlim(25.65)
plt.ylim(0,1)
plt.title('Funcion de Distribucion de Poisson')
plt.ylabel('Probailidad')
plt.xlabel('Valores')
plt.show()
```

# Distribucion normal

```python
mu, sigma = 0,1
limite_inferior = ss.norm(mu,sigma).ppf(0.001) # ppf entrega los percentiles (percent point function)
limite_superior = ss.norm(mu,sigma).ppf(0.999)
print(f"El limite inferior es {limite_inferior}")
print(f"El limite superior es {limite_superior}")
dist = np.linspace(limite_inferior,limite_superior, 100)

f_densidad = ss.norm(mu,sigma).pdf(dist)
f_dust_acumulada = ss.norm(mu,sigma).cdf(dist)

plt.figure(figsize = (8,6))
plt.plot(dist,f_densidad, label = "f dendidad")
plt.plot(dist,f_dust_acumulada, label = "f. dist acumulada")
plt.xlim(-4,4)
plt.grid(True)
plt.legend(shadow = True, frameon = True)
```

# Ejemplo

Se tiene una distrinucion normal X tal que mu = 1000 y sigma = 15 y se buscar determinar:

1) P(985 < X 1015)

2) p(-inf < X 1000)

3) P(x>= 990)

4) P(X < 1010)

```python
mu = 1000
sigma = 15

p1 = ss.norm(mu,sigma).cdf(1015) - ss.norm(mu,sigma).cdf(985)
print(f"La probabilidad es: {p1}")

p2 = ss.norm(mu,sigma).cdf(1000) - ss.norm(mu,sigma).cdf(-np.inf)
print(f"La probabilidad es de: {p2}")

p3 = 1 - ss.norm(mu,sigma).cdf(990)
print(f"la probabilidad es de: {p3}")

p4 = ss.norm(mu,sigma).cdf(1010)
print(f"la probabilidad es de: {p4}")
```
